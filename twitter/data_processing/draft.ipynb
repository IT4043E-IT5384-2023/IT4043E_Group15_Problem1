```python
import pandas as pd
from tqdm import tqdm
import json
import numpy as np

import pycountry_convert as pc
```

```python
profiles = pd.read_json("D:\Projects\Big data storage and processing\IT4043E_Group15_Problem1\output\profiles.json")
profiles = profiles.transpose()
idxs = [i for i in range(2150)]
profiles.index = idxs
profiles.head()
```

```python
users = []
for user in profiles["user"]:
    users.append(user)
users = pd.DataFrame.from_records(users, index=profiles.index)
users.head()
```

```python
unnecessary = ['profileImageUrl','profileBannerUrl','descriptionLinks','_type','id_str','url','created', 
               'rawDescription', 'protected', 'blueType', 'displayname']
users_dropped = users.drop(columns=unnecessary, axis=1)
users_dropped.head()
```

```python
users_dropped.info()
```

```python
tweets = []
for twt_lst in profiles['tweet']:
    for tweet in twt_lst:
        tweets.append(tweet)
tweets = pd.DataFrame.from_records(tweets)
tweets.head()
```

```python
tweets.info()
```

```python
tweets["user"] = tweets["user"].map(lambda x: x["id"])
tweets_dropped = tweets[["id", "user", "replyCount", "retweetCount", 
                         "likeCount", "quoteCount", "viewCount"]]
tweets_dropped.rename(columns={"user": "user_id"}, inplace=True)
tweets_dropped.head()
```

```python
with open('D:\Projects\Big data storage and processing\IT4043E_Group15_Problem1\output\location.json') as json_file:
    location_dict = json.load(json_file)
for k, v in location_dict.items():
    if v == "None":
        location_dict[k] = None
location_dict
```

```python
country = []
for i in range(users.shape[0]):
    k = users["location"][i]
    if k == "" or str(k) == "nan":
        country.append(None)
    else:
        v = location_dict[k]
        if isinstance(v, str):
            v = v.strip()
        country.append(v)
users["country"] = country
users["country"].unique()
```

```python
missing_cnt = {
    "Kosovo": "Europe",
    "Palestinian Territories": "Asia",
    "North Pole": "Arctic",
    "Ascension and Tristan da Cunha": "Africa",
    "Gornja Siga": "Europe"
}

def country_to_continent(country_name: str) -> str:
    try:
        country_alpha = pc.country_name_to_country_alpha2(country_name)
        country_continent_code = pc.country_alpha2_to_continent_code(country_alpha)
        country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
        return country_continent_name
    except:
        if country_name in ["Europe", "Africa", "Antarctica"]:
            return country_name
        return missing_cnt[country_name]

for i, x in enumerate(country):
    if x is not None:
        country[i] = country_to_continent(x)
    else:
        country[i] = None
```

```python
users_dropped["location"] = country
users_dropped["location"].unique()
```

```python
users_dropped.info()
```

# PREPROCESSING

```python
# drop null label + drop verified (full False) + encode blue
#users_processed = users_dropped[users_dropped["location"].notnull()]
users_processed = users_dropped.drop(columns=["verified"], axis=1)
users_processed.dropna(subset=users_processed.columns.difference(["location"]), inplace=True)

users_processed.rename(columns={"id": "user_id"}, inplace=True)
users_processed.info()
```

```python
# drop null + avg count by user
tweets_processed = tweets_dropped.drop(columns=["id"], axis=1)
tweets_processed = tweets_processed.dropna()
tweets_processed = tweets_processed.groupby(["user_id"]).mean().reset_index()
tweets_processed.info()
```

```python
# merge data
data = pd.merge(users_processed, tweets_processed, how='left', on=['user_id'])
data[tweets_processed.columns.difference(["user_id"])] = data[tweets_processed.columns.difference(["user_id"])].fillna(0)
data.info()
```

# ML

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
data = data[data["location"] != "Arctic"]
data["location"].unique()
data["blue"] = data["blue"].astype("object")
```

```python
isinstance(data["location"], pd.Series)
```

```python
cat_attrbs = data.select_dtypes(include="object").columns.tolist()[2:]
num_attrbs = data.select_dtypes(exclude="object").columns.tolist()[1:]
cat_attrbs
```

```python
fn = ColumnTransformer([
    ("std_scaler", StandardScaler(), num_attrbs),
    ("lbl_encoder", OrdinalEncoder(), cat_attrbs),
], remainder="drop")
pipe = Pipeline([
    ("preprocess_fn", fn),
])
X_train = pipe.fit_transform(data.drop(columns=["location"], axis=1))
lb_encoder = LabelEncoder()
y_train = lb_encoder.fit_transform(data["location"])
```

```python
LABEL_MAP = {
    'Africa': 0, 
    'Antarctica': 1, 
    'Asia': 2, 
    'Europe': 3, 
    'North America': 4,
    'Oceania': 5, 
    'South America': 6,
    'Arctic': 7
}
data["location"].map(LABEL_MAP).to_numpy()
```

```python
import numpy as np
np.unique(y_train, return_counts=True)
```

```python
lb_encoder.classes_
```

```python
rf_clf = RandomForestClassifier(n_estimators=95, 
                                max_depth=10,
                                random_state=42)
cross_val_score(rf_clf, X_train, y_train, scoring="roc_auc_ovo", cv=2, n_jobs=-1).mean()
```
